```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.random.rand(50)
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y, c='#223BDC',alpha=.9,marker='X')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("This is X vs Y")
plt.grid()
```

```python
x= np.linspace(1,10, 100)
y= np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y,'--r')
plt.xlabel("X-data")
plt.ylabel("Y-axis")
plt.title("This is sin wave graph")
plt.grid()
```

```python
x= ['a','b','c','d','e']
y=np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.bar(x,y,color='#7E10D5')
plt.xlabel("representing my categorical data")
plt.ylabel("representing my numerical data")
plt.title("This is my bar plot")
plt.grid()
```

```python
## plotting same Dataset in Horizontal order
plt.figure(figsize=(5,3))
plt.barh(x,y,color='#7E10D5')
plt.xlabel("representing my categorical data")
plt.ylabel("representing my numerical data")
plt.title("This is my bar plot")
plt.grid()
```

```python
data= [1,2,3,4,5,6,6,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize=(6,3))
plt.hist(data)
plt.xlabel("Representing Unique value")
plt.ylabel("Representing fequency")
plt.title("Hist Plot Example")
plt.grid()
plt.savefig("test.png")
```

```python
x= np.random.rand(50)
y= np.random.rand(50)
z= np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z,c='#9327E8',alpha=.8)
plt.xlabel("My x axis")
plt.ylabel("My Y-axis")
plt.show()
```
