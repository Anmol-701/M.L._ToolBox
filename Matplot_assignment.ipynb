### Problem_2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np           
np.random.seed(3)           
x=3+np.random.normal(0,2,50)            
y=3+np.random.normal(0,2,len(x))           
Note: Also add title xlabel and ylabel to the plot           

  - Scatter plot: Useful for exploring correlations or patterns between two variables, especially when data points have distinct characterstics.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))

# Create the scatter plot
plt.figure(figsize=(6,4))
plt.scatter(x,y,color='blue',alpha=0.7,edgecolors='black')

# Adding Labels and titles
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.title('Scatter plot')

# Show the plot
plt.grid(True)
plt.show()
```

### Problem_3: Why is the subplot() function used? Draw four line plots using the subplot() function.

  - subplot() function in matplotlib is used to create a multiple plots within single figure, allowing you to visualize and comparing different dataset as aspect of your data efficiency and compatibility 

### Use the following data
import numpy as np        
#for line 1:

```python
import matplotlib.pyplot as plt
import numpy as np
# Data for each line
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = x1  # Reuse x values for clarity
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = x1  # Reuse x values for clarity
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = x1  # Reuse x values for clarity
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create figure and subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 6))  # Create a 2x2 grid

# Plot lines on each subplot
axes[0, 0].plot(x1, y1, label='Line 1')
axes[0, 0].set_title('Line 1 Data')
axes[0, 0].legend()

axes[0, 1].plot(x2, y2, label='Line 2')
axes[0, 1].set_title('Line 2 Data')
axes[0, 1].legend()

axes[1, 0].plot(x3, y3, label='Line 3')
axes[1, 0].set_title('Line 3 Data')
axes[1, 0].legend()

axes[1, 1].plot(x4, y4, label='Line 4')
axes[1, 1].set_title('Line 4 Data')
axes[1, 1].legend()

# Adjust spacing and layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Problem_4: What is bar Plot? Why it is used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np               
company =np.array(['Apple','Microsoft','Google','AMD'])              
profit=np.array([3000,8000,1000,10000])                

### Bar plot
  -  Bar plot is graphical representaion of catagorical data using rectangular bars whose height and length are proportional to the value they represent
  - It is used to comparing numerical values across different categories, making them ideal for visualisation.

```python
import matplotlib.pyplot as plt
import numpy as np
company=np.array(['Apple','Microsoft','Google','AMD'])
profit=np.array([3000,8000,1000,10000])

# Bar plot verticle
plt.figure(figsize=(10,6))
plt.bar(company,profit,color='skyblue')
plt.title('Company profit (Verticle Bar Plot)')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Bar Plot Horizontal
plt.figure(figsize=(10,6))
plt.barh(company,profit,color='lightcoral')
plt.title('Company Profit (Horizontal Bar Plot)')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.gca().invert_yaxis()  # Invert y-axis to have highest profit at the top
plt.tight_layout()
plt.show()
```

### Problem_5: What is a box plot? Why is it used? using the following data plot a box plot.
box1=np.random.normal(100,10,200)             
box2=np.random.normal(90,20,200)           

  - It is also known as box-and-whisker plot, is a graphical representation of the distribution of set of numerical data
  - By using box plot you can easily the central tendency,spread and presence of outliers across different dataset by placing multiple box plots side by side
  - Also by looking at the shape and size of the box and whiskers, you can get a sense of the skewness and symmetry of the data distribution.

```python
import matplotlib.pyplot as plt

box1=np.random.normal(100,10,200)
box2=np.random.normal(90,20,200)
# Create box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], notch=True, vert=True, patch_artist=True, labels=['Box 1', 'Box 2'])
plt.title('Box Plots for Box 1 and Box 2')
plt.xlabel('Data Groups')
plt.ylabel('Values')
plt.grid(True)
plt.tight_layout()
plt.show()
```
