### Program_2: What are glyphs in Boken, and how can you add them to a Boken plot? Explain with an example.

  - Glyphs are the fundamental visual elements used to represent your data as shapes, lines, or symbols on a plot. These act as the building blocks to construct different visualization types like scatter plots, line charts, bar charts, and more.
  - To add them to Bokeh plot you have to import necessary libraries.
  - After that you have to prepare your data.(Organize data in separate list)
  - After that you have to create plot object.
  - Choose the right glyphs like circle(), line(), vbar()

```python
## Example
from bokeh.plotting import figure, show

x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

p = figure(title="Scatter Plot")
p.circle(x, y, size=10, color="red")  # Add circles with size and color

show(p)
```

### Problem_3: How can you customize the apperance of a Bokeh plot, including the axes, title, and legend?

```python
# Customizing the apperance of Bokeh Plot
from bokeh.plotting import figure,show
x=[1,2,3,4,5]
y=[3,4,5,6,7]
plot=figure(title="Scatter Plot")
plot.line(x,y,color='blue')
plot.xaxis.axis_label='X-Label'
plot.yaxis.axis_label='Y-Label'
show(plot)
```

### Problem_4: What is a Bokeh server, and how can you use it to create interactive plots that can be updates in real time?

  - A Bokeh server is a component of the Bokeh visualization library that unlocks the power of creating interactive web applications connected to real-time Python code.
  - While Bokeh can create static plots displayed in web browsers or saved as HTML files
### Creating Real time server with bokeh server
  - use the bokeh.io.show function with server=True argument
  - Create you plot logic as usual with bokeh plotting function.
  - Start the Bokeh server application using bokeh.io.run() function.
  - Open the provided URL in web browser and see the updates happening in real time.

### Problem_5: How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

```python
from flask import Flask, render_template_string
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

x = [1, 2, 3, 4, 5]
y = [3, 2, 7, 5, 4]
plot = figure(title="Scatter_plot")
plot.circle(x, y, size=20, color="green")
script, div = components(plot)

@app.route('/')
def index():
    html = f"""
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Bokeh Plot Example</title>
</head>
<body>
    {script}
    {div}
</body>
</html>
"""
    return render_template_string(html)

if __name__ == '__main__':
    app.run(debug=True)
```
